```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 리튬가격예측데이터
```

```python
import pandas as pd

df1 = pd.read_csv('./한국광해광업공단_리튬가격예측데이터.txt', encoding='utf-8', low_memory=False)
df1.tail()
```

```python
df1.info()
```

```python
df1['전망기간'] = pd.to_datetime(df1['전망기간'])
df1['전망기간'] = df1['전망기간'].dt.strftime('%y-%m-%d')
df1.tail()
```

```python
x = df1['전망기간']
y = df1['예측가격']
max_index = df1['예측가격'].idxmax()
min_index = df1['예측가격'].idxmin()
plt.figure(figsize=(16,10))
plt.plot(x,y)
plt.xlabel('Forecast period')
plt.ylabel('Predict price')
plt.tick_params(axis='x', direction='out', length=3, pad=3, labelsize=10, rotation=90, top=True)
plt.title('lithium price prediction')


# for i, value in enumerate(df1['예측가격']):
#     plt.text(df1['전망기간'][i], value + 10, str(value), ha='center')

plt.text(df1['전망기간'][max_index], df1['예측가격'][max_index]+10, f'Max: {df1["예측가격"][max_index]}', ha='center', color='red', fontweight='bold')
plt.text(df1['전망기간'][min_index], df1['예측가격'][min_index]+10, f'Min: {df1["예측가격"][min_index]}', ha='center', color='red', fontweight='bold')
    
plt.show()
```
